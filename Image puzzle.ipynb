```python
import numpy as np
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# from google.colab.patches import cv2_imshow
img = cv.imread("jigsaw.jpg")

crop_img1 = cv.flip(img[150:330, 515:700], 1)

img[150:330, 515:700] = crop_img1

crop_img2 = cv.flip(img[370:420, 370:795], 0)

img[370:420, 370:795] = crop_img2


b,g,r=cv.split(img[0:200, 0:190])
crop_img3=cv.merge([g,b,r])


crop_img4 = cv.flip(img[200:410, 0:190], 0)

img[200:400, 0:190] = crop_img3
img[0:210, 0:190] = crop_img4
final = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(final)
plt.axis('off') 
```
